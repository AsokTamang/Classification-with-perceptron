```python
m=30
X=np.random.randint(0,2,(2,m))  #random number between 0 inclusive and 2 exclusive and size of 2 rows and m columns
Y = np.logical_and(X[0]==0,X[1]==1).astype(int).reshape((1,m))  #output of logical AND operation between X and Y
X
```

```python
#activation function
def sigmoid_function(z):
    return 1 / (1 + np.exp(-z))
sigmoid_function(3.5)                
                
```

```python
def layer_size(X,Y):
    layer_X = X.shape[0]  #size of the input layer or say number of features in the input data sets
    layer_Y = Y.shape[0]  #size of the output
    return layer_X,layer_Y
n_x,n_y = layer_size(X,Y)   
```

```python
def initialize_parameters(n_x,n_y):  #this is the function which provides us the initial weight of the input and the bias for our model
    w = np.random.randn(n_y,n_x) * 0.01
    b = np.zeros((n_y,1))
    parameters = {
        'w':w
        ,'b':b
    }
    return parameters
parameters = initialize_parameters(n_x,n_y)    
    
```

```python
#now part of forward propagation
def forward_propagation(X,parameters):
    w = parameters['w']
    b = parameters['b']
    z = w @ X + b   #the shape of z will be n_y * m having n_y rows and m columns
    a = sigmoid_function(z) #as in the classification problem , we need the sigmoid of the predicted value
    return a
A = forward_propagation(X,parameters)    #this is the predicted sigmoid value 
```

```python
def cost_function(Y,A):
    m=Y.shape[1]
    logprobabs = -np.multiply(np.log(A),Y) - np.multiply(np.log(1-A),1-Y)  #formulae of log loss function
    log_loss = np.sum(logprobabs) / m    #summation of all 
    return log_loss
cost = cost_function(Y,A)    
cost    
```

```python
def back_propagation(A,Y,X):  #finding the partial derivatives for the gradient descent
    m = X.shape[1] #total number of samples
    dz = A - Y  #partial derivative of loss function with respect to Y-hat
    dlw = np.dot(dz,X.T) / m
    dlb = np.sum(dz,axis = 1) / m
    grads = {     
        'dlw':dlw,
        'dlb':dlb
    }
    return grads
grads = back_propagation(A,Y,X)
grads
```

```python
def update_parameters(parameters,grads,learning_rate = 1.2):
    w,b = parameters['w'],parameters['b']
    dlw,dlb = grads['dlw'],grads['dlb']
    w = w - learning_rate *dlw
    b = b - learning_rate * dlb
    parameters = {
        'w':w,
        'b':b
    }
    return parameters
updated_parameters = update_parameters(parameters,grads,learning_rate = 1.2)    
updated_parameters
```

```python
def nn_model(iterations,X,Y,learning_rate,print_cost=False): #our neural network model for classification
    n_x,n_y = layer_size(X,Y)
    parameters = initialize_parameters(n_x,n_y)   #initial parameters
    for i in range(iterations):
        A = forward_propagation(X,parameters)
        grads = back_propagation(A,Y,X)
        cost = cost_function(Y,A)
        if print_cost:
            print(f'The cost function at the iteration {i+1} is', cost)
        parameters = update_parameters(parameters,grads,learning_rate)
    return parameters
parameters = nn_model(50,X, Y,learning_rate=1.2,print_cost=True) 
parameters
    
        
        
```

```python
def plot_decision_boundary(X, Y, parameters):
    W = parameters["w"]
    b = parameters["b"]

    fig, ax = plt.subplots()
    plt.scatter(X[0, :], X[1, :], c=Y, cmap=colors.ListedColormap(['blue', 'red']));
    
    x_line = np.arange(np.min(X[0,:]),np.max(X[0,:])*1.1, 0.1)
    ax.plot(x_line, - W[0,0] / W[0,1] * x_line + -b[0,0] / W[0,1] , color="black")
    plt.plot()
    plt.show()
    
plot_decision_boundary(X, Y, parameters)
```
